```python
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\Lindsey\Desktop\Springboard\JobSearch_WordCloud\Job_Descriptions\Descriptions.txt", on_bad_lines='skip', encoding='utf-8')
```

```python
data = str(data)
```

```python
wordcloud = WordCloud().generate(data)
```

```python
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
from wordcloud import STOPWORDS
print(STOPWORDS) # a set of words
```

```python
STOPWORDS.update({"NaN", "ApiJect", "within", "highly", "great", "person", "responsibilities", "pages", "work" , "help", "maintain", "assist", "managers", "ability", "thinker", "include", "time", "critical", "real", "settle", "seeking" , "setting" , "doesn't", "beyond" , "strong", "execute", "efforts" , "transfer" , "create", "build" , "act", "organized","visitor" , "doesn" , "demonstrated" , "functional", "scope", "experience" , "th" , "able" , "customer" , "point", "working","prioritize", "onging", "various", "util","managing", "obsessed", "manage", "email", "campaigns" , "team" , "website", "digital", "marketing", "content", "position", "columns", "range", "landing", "ongoing" }) 

wordcloud = WordCloud(
        stopwords=STOPWORDS, max_words=100, width=1000, height=500
    ).generate(data)
```

```python
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
