### Encoding target feature to numeric

In order to visualize and model data, we need to convert the M and B diagnosis into binary of Malignant =1 & Benign = 0

```python
#Encoding M-> 1 and B-> 0
data['Y'] = data.diagnosis == 'M'
#Sanity check on the encoding
data[['Y', 'diagnosis']]
```

```python
#Second sanity check on the encoding to make sure it did not change diagnosis count. 
data.Y.value_counts()
```

```python
#Converting bool values to numeric for visualization and modeling
data.Y = data.Y.astype(int)
```

### What are the differences of features in each class?

#### Computing histograms grouped by classes to check on the differences in feature distribution.
#### Legend (0: Bening, 1: Malignant)

```python
#Plotting histograms with legend (0: Bening, 1: Malignant)
for column in cols:
    print(column)
    data.groupby(by='Y')[column].hist(bins=30, alpha=0.4, legend='Y')
    plt.show()
```

Histograms show features with greater differences between Malignant and Benign classes are:

- `radius_mean`
- `concave points_mean`
- `radius_worst`
- `perimeter_worst`
- `concave points_worst` 
- `area se`



This analysis suggests that the main differences between M / B biopsies could be mainly related to geometry of the sample.
